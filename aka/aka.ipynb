```python
import pandas as pd
import numpy as np
import os
import folium
import pgeocode
```

```python
 # Load the data
current_dir = os.getcwd()
print (f"Current directory: {current_dir}")
csv_dir = current_dir.replace('aka', 'datasets\\Kangaroo.csv')
df = pd.read_csv(csv_dir)
df.head(10)
```

```python
print(f"Shape = {df.shape}")
print(f"Columns : {df.columns}")
print(f"NAN values : \n{df.isna().sum()}")
print("Percentage of nan values : ")
print((df.isna().mean() * 100).round(2).astype(str) + ' %')
```

```python
# Number of rows and columns
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print(f"Columns: {df.columns.tolist()}")

# Check for missing values
print("Missing values:")
print(f"Price null : {df["price"].isnull().count()}")
print(f"Price null after drop : {df["price"].count()}")
df = df.dropna(subset=["price"])

print(f"habitableSurface null : {df["habitableSurface"].isnull().count()}")
print(f"facedeCount null : {df["facedeCount"].isnull().count()}")

df = df.fillna({
    "hasGarden": False,
    "hasTerrace": False,
    "hasBalcony": False,
    "hasCellar": False,
    "hasPool": False,
    "gardenSurface": 0,
    "terraceSurface": 0,
    "parkingCountOutdoor": 0,
    "parkingCountIndoor": 0,
    "hasFireplace": False,
})
```

```python
nomi = pgeocode.Nominatim('be')
# Get the coordinates 

df["Latitude"] = (nomi.query_postal_code(list(map(str,df["postCode"].tolist()))).latitude)
df["Longitude"] = (nomi.query_postal_code(list(map(str,df["postCode"].tolist()))).longitude)

df_map = df.drop_duplicates(subset=["postCode"])




df.head(25)
center_coordinates =nomi.query_postal_code(1000)  # Get the coordinates of the first row
# Create a map centered around the average latitude and longitude
map_center = df_map[['Latitude', 'Longitude']].mean()

map_plot = folium.Map(location=[center_coordinates.latitude,center_coordinates.longitude], zoom_start=8, control_scale=True)   

df_map = df.drop_duplicates(subset=["postCode"], keep='first') 
print(f"Total original postCode : {df.shape[0]}")
print(f"Total duplicated postCode : {df_map.shape[0]}")
print(f"Total original postCode : {df.duplicated(subset=["postCode"]).sum()}")
print(f"Total duplicated postCode : {df_map.duplicated(subset=["postCode"]).sum()}")

color = "yellow"
fill_color = "blue"
html_color = ""

for grp_name, df_grp in df_map.groupby('locality'):
    feature_group = folium.FeatureGroup(grp_name)
    for row in df_grp.itertuples():
        count = df[df['postCode'] == row.postCode].shape[0]  # Count occurrences of the postcode in the original DataFram
        meanPrice = df[df['postCode'] == row.postCode]['price'].mean()  # Calculate mean price for the postcode
        maxPrice = df[df['postCode'] == row.postCode]['price'].max()  # Calculate max price for the postcode
        minPrice = df[df['postCode'] == row.postCode]['price'].min()  # Calculate min price for the postcode
        if meanPrice<350000:
            color = "green"
            fill_color = "green"
            html_color = "#2B952A"
        elif 350000<meanPrice<500000:
            color = "orange"
            fill_color = "orange"
            html_color = "#FFA501"
            
        elif 500000<meanPrice:
            color = "red"
            fill_color = "red"
            html_color = "#FF0000"

        html = f"""
                <h1> <b>{row.postCode} : {grp_name}</b></h1><br>
                <h2><b> Total sells : {count}</b></h2><br>
                <h2 style='color:{html_color};'> <b> Mean Price : {meanPrice:.0f} €</b></h2><br>
                <h2> <b> Max Price : {maxPrice:.0f} €</b></h2><br>
                <h2> <b> Min Price : {minPrice:.0f} €</b></h2><br>
                """
        popup = folium.Popup(html, max_width=350)


        

        folium.CircleMarker(
            location=[row.Latitude, row.Longitude],
            popup=popup,
            radius=5, # define how big you want the circle markers to be
            color=color,
            fill=True,
            fill_color=fill_color,
            fill_opacity=0.6).add_to(feature_group),
    feature_group.add_to(map_plot)

# Add LayerControl to toggle FeatureGroups
folium.LayerControl().add_to(map_plot)


# Save the map to an HTML file
map_plot.save("map_plot.html")
map_plot
```

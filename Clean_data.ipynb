```python
import numpy as np
import pandas as pd

data = './datasets/Kangaroo.csv'
data_set = pd.read_csv(data, sep = ',')
df = pd.DataFrame(data_set)

print(df)
```

```python
na_values = df.isnull().sum()
missing_percent = (na_values/ len(df)) * 100
print (missing_percent.round(2))
```

Step 1 : Data Cleaning

```python
# Remove non relevant columns
df= data_set.drop(columns=['url','monthlyCost','accessibleDisabledPeople','gardenOrientation','terraceOrientation'])
print(df)
```

```python
duplicates = data_set.duplicated()
print(duplicates.count)
```

```python
#Replace Missing_value to "0"
df = df.fillna(0)
print(df)
```

```python
#Replace "True" by 1
df = df.replace(True, 1)
print(df)
```

```python
# Delete row when the price == 0
df = df[df['price'] != 0]
print(df)
```

```python
df.to_csv("my_clean_data.csv", index=False, encoding='utf-8')
```
